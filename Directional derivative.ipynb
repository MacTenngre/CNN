## 1. 方向导数 （Directional derivative）
已知旋转抛物面：$z = f(x, y)$  
在点 $(x_{0}, y_{0})$ 处沿单位向量$\vec{u}=(u_{1}, u_{2})$的方向导数可以定义为：  
  
$$D_{u}f(x_{0}, y_{0}) = \lim_{h\rightarrow0} \frac{f(x_{0} + hu_{1}, f(y_{0} + hu_{2}) - f(x_{0}, y_{0})}{h}$$  
  
沿着$x$轴$(1, 0)\implies$对$x$偏导数  
$$D_{u}f(x_{0}, y_{0}) = \lim_{h\rightarrow0} \frac{f(x_{0} + h, f(y_{0}) - f(x_{0}, y_{0})}{h} = f_{x}(x_{0}, y_{0})$$  
  
沿着$y$轴$(0, 1)\implies$对$y$偏导数    
$$D_{u}f(x_{0}, y_{0}) = \lim_{h\rightarrow0} \frac{f(x_{0}, f(y_{0} + h) - f(x_{0}, y_{0})}{h} = f_{y}(x_{0}, y_{0})$$  

由此可以将方向导数转化为对分量的偏导数与单位向量的内积：
$$D_{u}f(x_{0}, y_{0}) = <f_{x}(x_{0}, y_{0}),  f_{y}(x_{0}, y_{0})> \cdot <u_{1}, u_{2}>$$  

$$D_{u} = \lim_{h\rightarrow0} \frac{\overbrace{f(x_{0} + hu_{1}, y_{0} + hu_{2})}^{set\ as\ a\ function\ with\ h} - f(x_{0}, y_{0})}{h}$$  
$$F(h) = f(x_{0} + hu_{1}, y_{0} + hu_{2})$$  
对$F(h)$求全微分 $$dF = F_{x}dx + F_{y}dy$$  
$$\frac{dF}{dh} = F_{x}\frac{dx}{dh} + F_{y}\frac{dy}{dh} = <F_{x}, F_{y}> \cdot <u_{1}, u_{2}>$$  

> * $F_{x}$ 为当$y$固定时，自变量$x$对应于$x$轴的斜率。$F_{y}$同理。  
> * 详细数学推导请参阅《Calculus》James Stewart 6th edition chapter14.6

## 2. 方向导数实操 

已知
* $f(x, y) = x^{2} + y^{2} + 4$
* $(1, 0) 方向为 <\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2}>$ 的方向导数

$$<f_{x}, f_{y}> \cdot <\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2}> = <2x,2y>|_{(1, 0)} \cdot <\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2}>$$
$$<2, 0> \cdot <\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2}> = \sqrt{2}$$  

在抛物曲面 $f(x, y) = x^{2} + y^{2} + 4$ 上的点 $(1, 0, 5)$，沿着 $<\frac{\sqrt{2}}{2}, \frac{\sqrt{2}}{2}>$ 方向的斜率为$\sqrt{2}$

## 3. 梯度

$D_{u}f=\overbrace{<f_{x}, f_{y}>}^{defined\ as\ grad} \cdot <u_{1}, u_{2}>$  
$D_{u}f = |\nabla f| \cos\theta$  
$D_{u}f = 2xi + 2yj$
$|\nabla f(1, 0)| = \sqrt{(2x)^{2} +(2y)^{2}} = 2$

## 4. 可视化 based on matplotlib 及 python 

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import numpy as np
%matplotlib inline
plt.rcParams['figure.figsize'] = (18.0, 16.0)
```

```python
fig = plt.figure()
ax = fig.gca(projection='3d')

# Make data. 
X = np.arange(-1, 3, 0.05)
Y = np.arange(-1, 3, 0.05)
X, Y = np.meshgrid(X, Y)
Z = (X**2 + Y**2) + 4
# Z = np.sin(Z)
# contour line
cset = ax.contour(X, Y, Z, zdir='z', stride=1, offset=-0, cmap=cm.coolwarm)

# Plot the surface. 二次抛物曲面
ax.plot_surface(X, Y, Z, rstride=8, cstride=8, alpha=1)
#---------------------

# 零平面
x = np.arange(-3, 4, 0.05)
y = np.arange(-3, 4, 0.05)
x, y = np.meshgrid(x, y)
z = x*0 + y*0
ax.plot_surface(x, y, z, rstride=8, cstride=8, alpha=0.3)
#----------------------

# xy平面上两点， 及直线y=x-1
x1 = np.linspace(1, 2, 20)
y1 = x1 -1
z1 = x1 * 0
ax.scatter(1, 0, 0)
ax.scatter(2, 1, 9)

ax.plot(x1, y1, z1)
#----------------------

# 切平面
z2 = np.linspace(0, 5, 10)
z3 = np.linspace(0, 9, 20)
ax.scatter(2, 1, z3)
ax.scatter(1, 0, z2)
z3 = 2*x + 3
ax.plot_surface(x, y, z3, rstride=8, cstride=8, alpha=0.3)
#-----------------------

#  切平面上直线
x4 = np.linspace(0, 2.5, 10)
y4 = x4 - 1
z4 = 2*x4 + 3
ax.plot(x4, y4, z4)

# 梯度
x5 = np.linspace(1, 1.6, 10)
y5 = x5 * 0
z5 = x5 * 0
ax.plot(x5, y5, z5)

#
#ax.zaxis.set_major_locator(LinearLocator(10))
#ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

# define x, y, z axis
ax.set_xlabel('X')
ax.set_xlim(-1.5, 3.5)
ax.set_ylabel('Y')
ax.set_ylim(-1.5, 3.5)
ax.set_zlabel('Z')
ax.set_zlim(0.2, 10)

#
plt.show()
```
